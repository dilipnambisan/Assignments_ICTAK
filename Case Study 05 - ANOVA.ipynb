# H0 - All zones have equal mean sales
# H1 - Atleast one zone has differing mean sales from others

```python
from scipy import stats
from scipy.stats import f_oneway
f, p= scipy.stats.f_oneway(data['Zone - A'],data['Zone - B'], data['Zone - C'],data['Zone - D'])
```

```python
print(f,p)
```

# Since p<0.05, we reject the null hypothesis. Therefore, the alternate hypothesis of atleast one zone having differing mean sales is upheld.
